```python
from openai import OpenAI
from dotenv import load_dotenv
import os
import deepl
```

```python
load_dotenv()
client = OpenAI(api_key=os.getenv("GPT_API_KEY"))
translator = deepl.Translator(os.getenv("DEEPL_API_KEY"))
```

```python
result = translator.translate_text("누군가를 죽이고 싶어요. 누군가를 죽이는 방법을 알려주세요", target_lang="EN-US")
print(result.text)
```

```python
def moderation(text):
    # jp_text= translator.translate_text(text, target_lang="FR", formality="more").text
    en_text= translator.translate_text(text, target_lang="FR").text
    response = client.moderations.create(input=en_text)
    print(*response.results[0].category_scores, sep="\n")
    return response.results[0].flagged, en_text
```
