```python
# 일기 텍스트 임베딩 생성
diary_embedding_response = client.embeddings.create(
    input=texts,
    model="text-embedding-3-large"
)

# 해시태그 임베딩 생성
hashtag_embedding_response = client.embeddings.create(
    input=hashtags,
    model="text-embedding-3-large"
)
```

```python
hashtag_embedding_response.data[0]
```

```python
# 임베딩 추출
diary_embedding = np.array(diary_embedding_response.data[0].embedding)
hashtag_embeddings = np.array([item.embedding for item in hashtag_embedding_response.data])
```

```python
similarities = cosine_similarity([diary_embedding], hashtag_embeddings)[0]
```

```python
similarities
```

```python
# 유사도가 가장 높은 3개의 해시태그 추천
top_indices = similarities.argsort()[-3:][::-1]
recommended_hashtags = [hashtags[i] for i in top_indices]

print("추천 해시태그:", recommended_hashtags)
```
